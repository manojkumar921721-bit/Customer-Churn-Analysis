# Customer Churn Analysis

## Objective

The objective of this project is to analyze customer churn patterns and identify the factors contributing to customer attrition. The insights generated can help businesses improve customer retention and reduce churn rates.

### Tools Used
- Python
- Pandas
- NumPy
- Matplotlib
- Seaborn

### Dataset
Telco Customer Churn Dataset

# Import Libraries

# Load Dataset

```python
import pandas as pd

df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

# Dataset Overview

```python
df.shape
df.info()
df.describe()
```

# Data Cleaning

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Churn Distribution

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Churn', data=df)

plt.title('Customer Churn Distribution')
plt.show()
```

### Observation

Most customers remain with the company, while a significant portion have churned.

## Gender vs Churn Analysis

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x='gender',
    hue='Churn',
    data=df
)

plt.title('Gender vs Churn')
plt.show()
```

### Observation

Churn rates appear relatively similar between male and female customers.

## Contract Type vs Churn Analysis

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x='Contract',
    hue='Churn',
    data=df
)

plt.title('Contract Type vs Churn')
plt.xticks(rotation=15)
plt.show()
```

### Observation

Customers with month-to-month contracts have substantially higher churn compared to customers with one-year or two-year contracts.

## Monthly Charges vs Churn

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Churn',
    y='MonthlyCharges',
    data=df
)

plt.title('Monthly Charges vs Churn')
plt.show()
```

### Observation

Customers who churn tend to have higher monthly charges.

## Tenure Analysis

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x='tenure',
    hue='Churn',
    bins=30
)

plt.title('Tenure Distribution by Churn')
plt.show()
```

### Observation

Customers with shorter tenure are more likely to churn.

## Correlation Analysis

```python
numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(8,5))

sns.heatmap(
    numeric_df.corr(),
    annot=True
)

plt.title('Correlation Heatmap')
plt.show()
```

# Key Insights

1. Month-to-month contracts have the highest churn rate.

2. Customers with higher monthly charges are more likely to leave.

3. Customers with shorter tenure are at greater risk of churn.

4. Long-term contracts improve customer retention.

5. Gender has little impact on churn behavior.

# Key Insights

### Insight 1
Customers with month-to-month contracts exhibit the highest churn rate.

### Insight 2
Customers paying higher monthly charges are more likely to churn.

### Insight 3
Long-term customers tend to remain loyal and show lower churn rates.

### Insight 4
Contract duration plays a significant role in customer retention.

# Conclusion

This analysis identified key factors influencing customer churn, including contract type, tenure, and monthly charges. Businesses can improve customer retention by encouraging long-term contracts and focusing on high-risk customer segments.
